```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_csv('scores.csv')
df
```

```python
def comfort_label(score):
    if score >= 4:
        return 1  # 1 - комфортно
    else:
        return 0  # 0 - некомфортно
```

```python
for col in df.columns[1:]:
    df[f'{col}_comfortable'] = df[col].apply(comfort_label)
```

```python
df
```

```python
df_long = pd.melt(df, id_vars=['id'],
                  value_vars=[f's{i}_comfortable' for i in range(1, 10)],
                  var_name='element', value_name='comfortable')
```

```python
X = df_long[['element']]
X = pd.get_dummies(X, columns=['element'], drop_first=True)
y = df_long['comfortable']
```

```python
X = sm.add_constant(X)
```

```python
logit_model = sm.Logit(y, X.astype(float))
result = logit_model.fit()
```

```python
print(result.summary())
```
